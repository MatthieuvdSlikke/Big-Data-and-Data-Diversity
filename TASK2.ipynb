```python
import numpy as np
np.random.seed(0)
df = pd.DataFrame(np.random.randn(313,4), columns=["population_density", "dwellings_density", "balance_services", "transport_density"])
df
```

```python
def z_score(df):
        df.columns = [x + "_zscore" for x in df.columns.tolist()]
        return ((df - df.mean())/df.std(ddof=0))
```

```python
z_score(df)
```

```python
import matplotlib.pyplot as plt
Data=pd.read_csv('CensusStats.csv')
```

```python
median_annual_household_income = [row[median_annual_household_income] for row in Data]
walkability = [row[walkability] for row in Data]
plt.scatter(median_annual_household_income, walkability)
plt.title('median_annual_household_income vs walkability')
plt.xlabel('median_annual_household_income')
plt.ylabel('walkability')
plt.axis([-1,25,-1,25])
plt.show()
```
